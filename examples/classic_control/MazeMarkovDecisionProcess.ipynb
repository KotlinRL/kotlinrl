## Markov Decision Process using KotlinRL Maze Env

```kotlin
USE {
    repositories {
        mavenCentral()
        maven("https://central.sonatype.com/repository/maven-snapshots/")
    }
    dependencies {
        implementation("io.github.kotlinrl:envs:0.1.0-develop-SNAPSHOT")
    }
}
//%use dataframe
//%use kandy
```

```kotlin
import io.github.kotlinrl.core.agent.*
import io.github.kotlinrl.core.learn.*
import io.github.kotlinrl.core.learn.mcc.*
import io.github.kotlinrl.core.policy.*
import io.github.kotlinrl.core.space.*
import io.github.kotlinrl.core.train.*
import io.github.kotlinrl.core.wrapper.*
import io.github.kotlinrl.core.env.*
import io.github.kotlinrl.envs.*
```

```kotlin
val env = Maze()
var (state, _) = env.reset()
var done = false
val gamma = 0.99
var G_0 = 0.0
var t = 0
while (!done) {
    val action = env.actionSpace.sample()
    var (_, reward, terminated, truncated, _) = env.step(action)
    done = terminated || truncated
    G_0 += gamma.pow(t) * reward
    t += 1
}
env.close()
println("""
It took us $t steps to reach the goal.
and each reward r(s, a)**-1, so the return amounts to ${G_0}
""")
```
