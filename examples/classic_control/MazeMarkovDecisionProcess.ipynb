## Markov Decision Process using KotlinRL Maze Env

```kotlin
USE {
    repositories {
        mavenCentral()
        maven("https://central.sonatype.com/repository/maven-snapshots/")
    }
    dependencies {
        implementation("io.github.kotlinrl:envs:0.1.0-develop-SNAPSHOT")
    }
}
//%use dataframe
//%use kandy
```

```kotlin
import io.github.kotlinrl.core.agent.*
import io.github.kotlinrl.core.learn.*
import io.github.kotlinrl.core.learn.mcc.*
import io.github.kotlinrl.core.policy.*
import io.github.kotlinrl.core.space.*
import io.github.kotlinrl.core.train.*
import io.github.kotlinrl.core.wrapper.*
import io.github.kotlinrl.core.env.*
import io.github.kotlinrl.envs.*
```

```kotlin
val env = RecordVideo(Maze(metadata = emptyMap(), render = true), folder = "videos/maze_mdp", fps = 5)
```

```kotlin
var (obs, _) = env.reset()
val frames = mutableListOf<Rendering.RenderFrame>()
frames + env.render() as Rendering.RenderFrame
for(i in 0 until 4) {
    env.step(Maze.Action.DOWN.value)
    frames + env.render() as Rendering.RenderFrame
}
for(i in 0 until 3) {
    env.step(Maze.Action.RIGHT.value)
    frames + env.render() as Rendering.RenderFrame
}
env.step(Maze.Action.UP.value)
frames + env.render() as Rendering.RenderFrame
for(i in 0 until 2) {
    env.step(Maze.Action.RIGHT.value)
}
var (_, reward, terminated, truncated, _) = env.step(Maze.Action.DOWN.value)
frames + env.render() as Rendering.RenderFrame
println("Reward: $reward, Done: ${terminated || truncated}")
env.close()
env.displayVideo(1)
```
